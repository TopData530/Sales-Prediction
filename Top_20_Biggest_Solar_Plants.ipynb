```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('/content/20 Largest Solar Power Plants.xlsx')
df.head()
```

```python
df.duplicated().sum()
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
plt.style.use('seaborn');
sns.barplot(x = 'Acres', y = 'Name', edgecolor='black', data=df);
plt.xlabel('Acres', fontsize=20);
plt.ylabel('Name Of Plant', fontsize=20);
plt.title('Largest Solar Plants in the World', fontsize=20);
```

```python
plt.style.use('seaborn')
df['Country'].hist(edgecolor='black')
set_style('darkgrid')
plt.xticks(rotation=45);
plt.xlabel('Countries', fontsize=20);
plt.ylabel('Number of Solar Plants', fontsize=20);
plt.title('Number of Plants by Countries', fontsize=20);
```
